```python
## Mathematical Operation
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

### Union
union_set = set1.union(set2)
print(union_set)

## Intersection
intersection_set = set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2)
print(set1)
```

```python
# sets
set1 = {1, 2, 3, 4, 5, 6}
set2 = {4, 5, 6, 7, 8, 9}

## Difference
print(set1.difference(set2))   # elements in set1 but not in set2
print(set2.difference(set1))   # elements in set2 but not in set1
```

```python
## Sets Methods
set1 = {1, 2, 3, 4, 5}
set2 = {3, 4, 5}

## is subset
print(set1.issubset(set2))    # is set1 ⊆ set2 ?
print(set2.issubset(set1))    # is set2 ⊆ set1 ?

## is superset
print(set1.issuperset(set2))  # is set1 ⊇ set2 ?
print(set2.issuperset(set1))  # is set2 ⊇ set1 ?
```

```python
### Counting unique words in text

text = "In this tutorial we are discussing about sets"
words = text.split()

# convert list of words to set to get unique words
unique_words = set(words)
print(unique_words)
print(len(unique_words))
```

```python
# create a set
my_set = {1, 2, 3, 4, 5, 7}

# discard method
my_set.discard(11)        # 11 is not in the set, so nothing happens
print(my_set)             # output: {1, 2, 3, 4, 5, 7}

# pop method
removed_element = my_set.pop()
print(removed_element)    # prints some element removed from the set
print(my_set)             # remaining elements after pop
```

```python
student = {"name": "Mahadev", "age": 45, "grade": 24}
print(student)
print(type(student))

# Single key is always used (duplicate key example)
student = {"name": "Mahadev", "age": 45, "name": 24}
print(student)
```

```python
student = {"name": "Mahadev", "age": 45, "grade": 24}

## Accessing dictionary elements
print(student["grade"])
print(student["age"])

## Accessing using get() method
print(student.get("grade"))
print(student.get("last_name"))
print(student.get("last_name", "Not Available"))
```

```python
### Iterating Over Dictionaries
# You can use loops to iterate over dictionaries, keys, values, or items

student = {"name": "Mahadev", "age": 33, "grade": "A", "address": "India"}

## Iterating over keys
for key in student.keys():
    print(key)
```

```python
## Nested Dictionaries
students = {
    "student1": {"name": "Mahadev", "age": 45},
    "student2": {"name": "Peter", "age": 35}
}

print(students)
```

```python
students = {
    "student1": {"name": "Mahadev", "age": 45},
    "student2": {"name": "Peter", "age": 35}
}

## Iterating over nested dictionaries
for student_id, student_info in students.items():
    print(f"{student_id}: {student_info}")
    for key, value in student_info.items():
        print(f"{key}: {value}")
```
